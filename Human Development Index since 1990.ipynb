# Data Cleaning

## First analysis

- The values in the dataframe must be slightly changed.
1. The column HDI Rank can be dropped, to avoid problem during analysis.
2. Problem with the values, because it's objects and not float. 
3. The HDI rank must be in the last place: it's not relevant for the moment.
4. Some values doesn't exist, so it must be replaced by NaN.

To avoid problems with the DataFrame if there is a problem, a backup copy has to be done.

```python
df = hdi.copy()
```

```python
# Drop row HDI Rank
df.drop('HDI Rank', axis=1, inplace=True)
```

```python
# Replace the .. which are aliases for no value and spaces 
df=df.replace('..', '').replace(' ', '')
```

```python
# Select the columns to transform it into float excluding the Country column
cols = df.columns[df.dtypes.eq('object')]
cols=cols.drop('Country')
```

```python
df[cols]=df[cols].apply(pd.to_numeric, downcast='float')
df.head()
```

## Divide the DataFrame in 2

```python
df1=df.iloc[1:190]
df2=df.iloc[190:207]
```

```python
df1
```

```python
list_na={i:df1.iloc[i].isna().sum()/df1.shape[1] for i in range(len(df1.index)) if df1.iloc[i].isna().sum()/df1.shape[1]}
```

```python
sorted_na = sorted(list_na.items(), key=lambda kv: kv[1])
```

```python
sorted_na
```

```python
#interpolation works only with numeric columns. So, select numeric columns first
data_numeric=df1._get_numeric_data().copy()

#select numeric columns with missing data
cols_interp=data_numeric.columns[data_numeric.isnull().sum()>0]
cols_interp
```

```python
#interpolate them
data_numeric[cols_interp]=data_numeric[cols_interp].interpolate(method='linear')

data_numeric.isna().sum().sum()

df1[cols_interp]=data_numeric[cols_interp].interpolate(method='linear')

df1.columns[df1.isna().sum()>0]
```
